# NORMALIZATION

```python
heart_df['trestbps']=np.log(heart_df['trestbps'])
heart_df['chol']=np.log(heart_df['chol'])
heart_df['thalach']=np.log(heart_df['thalach'])

np.var(heart_df[["trestbps",'chol','thalach']])
```

```python
heart_df.isnull().sum()
```

```python
x=heart_df.drop('target',axis=1)    
y=heart_df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x,y, test_size=0.3,random_state=25)
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(heart_df.drop('target', axis=1))
```

```python
x.info()
```

# LOGISTIC REGRESSION

```python
accuracies={}

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
lr = LogisticRegression(penalty='l2')
lr.fit(x_train,y_train)

pred_y = lr.predict(x_test)

acc=accuracy_score(y_test,pred_y )
accuracies['LR']=acc*100
print("Training accuracy score of the model is:",accuracy_score(y_train, lr.predict(x_train))*100,"%")
print("Testing accuracy score of the model is:",accuracy_score(y_test,pred_y )*100,"%")


print("Confusion matrix of the model: ",confusion_matrix(y_test,pred_y))

print("Classification Report: ",classification_report(y_test,pred_y))
```

 # k-Nearest Neighbors (k-NN)

```python
knn=KNeighborsClassifier(n_neighbors=8)

knn.fit(x_train,y_train)

pred_y1 = knn.predict(x_test)

acc1=accuracy_score(y_test,pred_y1)
accuracies['KNN']=acc1*100

print("Training accuracy score of the model is:",accuracy_score(y_train, knn.predict(x_train))*100,"%")
print("Testing accuracy score of the model is:",accuracy_score(y_test,pred_y1)*100,"%")



print("Confusion matrix of the model",confusion_matrix(y_test,pred_y1))

print("Classification Report",classification_report(y_test,pred_y1))
```

# Support Vector Machines (SVM)

```python
from sklearn.svm import SVC

svc = SVC(probability=True)
svc.fit(x_train, y_train)

pred_y2 = svc.predict(x_test)

acc2=accuracy_score(y_test,pred_y2)
accuracies['SVM']=acc2*100

print("Training accuracy score of the model is:",accuracy_score(y_train, svc.predict(x_train))*100,"%")
print("Testing accuracy score of the model is:",accuracy_score(y_test,pred_y2)*100,"%")


print("Confusion matrix of the model",confusion_matrix(y_test,pred_y2))

print("Classification Report",classification_report(y_test,pred_y2))
```

# Decision Trees:


```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)

pred_y3 = dtc.predict(x_test)

acc3=accuracy_score(y_test,pred_y3)
accuracies['DT']=acc3*100

print("Training accuracy score of the model is:",accuracy_score(y_train, dtc.predict(x_train))*100,"%")
print("Testing accuracy score of the model is:",accuracy_score(y_test,pred_y3)*100,"%")


print("Confusion matrix of the model",confusion_matrix(y_test,pred_y3))

print("Classification Report",classification_report(y_test,pred_y3))
```

```python
from sklearn.model_selection import GridSearchCV

grid_params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : range(2, 32, 1),
    'min_samples_leaf' : range(1, 10, 1),
    'min_samples_split' : range(2, 10, 1),
    'splitter' : ['best', 'random']
}

grid_search = GridSearchCV(dtc, grid_params, cv = 10, n_jobs = -1, verbose = 1)
grid_search.fit(x_train, y_train)
```
